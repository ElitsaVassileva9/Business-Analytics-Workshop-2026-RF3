```python
#We need an extremely simple baseline that requires minimal modeling assumptions. 
#We need it to show that our other “smarter” models add value. 
#The simplest prediction of trolleys in period t is the number of trolleys in previous period t-1.
#Trolleys today = Trolleys yesterday (lag1): Yt=Yt-1
#Trolleys today = Trolleys 1 week ago (lag7): Yt=Yt-7
```

```python
#Holidays Only Lag1.
```

```python
holidays_only_df_baseline = holidays_only_df.sort_values(
    ["logistieke location id", "crop_type", "fpc_index", "date"]
)

# Create lag1
holidays_only_df_baseline["trolleys_lag1"] = (
    holidays_only_df_baseline
    .groupby(["logistieke location id", "crop_type", "fpc_index"])["trolleys"]
    .shift(1)
)

# Impute NaNs with median of that lag column 
holidays_only_df_baseline["trolleys_lag1"] = (
    holidays_only_df_baseline
    .groupby(["logistieke location id", "crop_type", "fpc_index"])["trolleys_lag1"]
    .transform(lambda x: x.fillna(x.median()))
)

holidays_only_df_baseline.head()
```

```python
#Holidays Only Lag7.
```

```python
# Create lag7
holidays_only_df_baseline["trolleys_lag7"] = (
    holidays_only_df_baseline
    .groupby(["logistieke location id", "crop_type", "fpc_index"])["trolleys"]
    .shift(7)
)

# Impute NaNs with median of that lag column 
holidays_only_df_baseline["trolleys_lag7"] = (
    holidays_only_df_baseline
    .groupby(["logistieke location id", "crop_type", "fpc_index"])["trolleys_lag7"]
    .transform(lambda x: x.fillna(x.median()))
)

holidays_only_df_baseline.head()
```

```python
import numpy as np

def mape(y_true, y_pred):
    mask = (y_true != 0) & (~y_true.isna()) & (~y_pred.isna())
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100
def mae(y_true, y_pred):
    mask = (~y_true.isna()) & (~y_pred.isna())
    return np.mean(np.abs(y_true[mask] - y_pred[mask]))
```

```python
#fpc-level MAPE for Holidays Only.
```

```python
holidays_only_mape_lag1 = mape(
    holidays_only_df_baseline["trolleys"],
    holidays_only_df_baseline["trolleys_lag1"]
)

holidays_only_mape_lag7 = mape(
    holidays_only_df_baseline["trolleys"],
    holidays_only_df_baseline["trolleys_lag7"]
)

print(f"[FPC | Holidays-only] Lag-1 baseline MAPE: {holidays_only_mape_lag1:.2f}%")
print(f"[FPC | Holidays-only] Lag-7 baseline MAPE: {holidays_only_mape_lag7:.2f}%")
```

```python
#fpc-level MAE for Holidays Only.
```

```python
holidays_only_mae_lag1 = mae(
    holidays_only_df_baseline["trolleys"],
    holidays_only_df_baseline["trolleys_lag1"]
)

holidays_only_mae_lag7 = mae(
    holidays_only_df_baseline["trolleys"],
    holidays_only_df_baseline["trolleys_lag7"]
)

print(f"[FPC | Holidays-only] Lag-1 baseline MAE:  {holidays_only_mae_lag1:.2f}")
print(f"[FPC | Holidays-only] Lag-7 baseline MAE:  {holidays_only_mae_lag7:.2f}")
```

```python
#On holiday days, the lag-7 baseline performs very poorly (MAPE ≈ 179%), 
#indicating that weekly seasonality breaks down during holidays. 
#The lag-1 baseline performs better (MAPE ≈ 40%) but still lacks predictive accuracy. 
#This highlights the need for models that explicitly account for holiday effects.”
```

```python
holidays_crop_level = (
    holidays_only_df_baseline
    .groupby(["date", "logistieke location id", "crop_type"])
    .agg(
        trolleys_true=("trolleys", "sum"),
        trolleys_pred_lag1=("trolleys_lag1", "sum"),
        trolleys_pred_lag7=("trolleys_lag7", "sum"),
    )
    .reset_index()
)
```
